```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
values = [['32x32(No scaling-SVM(Poly))',80],['32x32(No scaling-SVM(rbf))',84],['32x32(Normalizer-SVM(Poly))',81],
          ['32x32(Normalizer-SVM(RBF))',83],['64x64(No scaling-SVM(poly))',81],['64x64(No scaling-SVM(RBF))',85],
          ['64x64(Normalizer-SVM(poly))',83],['64x64(Normalizer-SVM(RBF))',84],['128x128(Normalizer-SVM(RBF))',82],
          ['128x128(No scaling-SVM(RBF))',85]]
```

```python
notes = pd.DataFrame(values,columns=['model','accuracy'])
```

```python
notes
```

```python
sns.lineplot(data = notes,x='model',y='accuracy')
plt.xticks(rotation = 90)
plt.title('Accuracies of all best models')
plt.show()
```

```python
f1_values = [['32x32(No scaling-SVM(Poly))',75],['32x32(No scaling-SVM(rbf))',79],['32x32(Normalizer-SVM(Poly))',80],
          ['32x32(Normalizer-SVM(RBF))',82],['64x64(No scaling-SVM(poly))',81],['64x64(No scaling-SVM(RBF))',85],
          ['64x64(Normalizer-SVM(poly))',83],['64x64(Normalizer-SVM(RBF))',83],['128x128(Normalizer-SVM(RBF))',82],
          ['128x128(No scaling-SVM(RBF))',84]]
notes = pd.DataFrame(f1_values,columns=['model','f1-score'])
```

```python
sns.barplot(data = notes,x='model',y='f1-score')
plt.xticks(rotation = 90)
plt.title('F1-score of all best models')
plt.show()
```
